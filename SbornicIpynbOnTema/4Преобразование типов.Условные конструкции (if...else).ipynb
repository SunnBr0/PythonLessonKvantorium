3. Операторы сравнения
Как правило, в качестве проверяемого условия используется результат вычисления одного из следующих операторов сравнения:

<
Меньше — условие верно, если первый операнд меньше второго.
>
Больше — условие верно, если первый операнд больше второго.
<=
Меньше или равно.
>=
Больше или равно.
==
Равенство. Условие верно, если два операнда равны.
!=
Неравенство. Условие верно, если два операнда неравны.
Например, условие (x * x < 1000) означает “значение x * x меньше 1000”, а условие (2 * x != y) означает “удвоенное значение переменной x не равно значению переменной y”.

Операторы сравнения в Питоне можно объединять в цепочки (в отличии от большинства других языков программирования, где для этого нужно использовать логические связки), например, a == b == c или 1 <= x <= 10.

Логическое И является бинарным оператором (то есть оператором с двумя операндами: левым и правым) и имеет вид and. Оператор and возвращает True тогда и только тогда, когда оба его операнда имеют значение True.

Логическое ИЛИ является бинарным оператором и возвращает True тогда и только тогда, когда хотя бы один операнд равен True. Оператор “логическое ИЛИ” имеет вид or.

Логическое НЕ (отрицание) является унарным (то есть с одним операндом) оператором и имеет вид not, за которым следует единственный операнд. Логическое НЕ возвращает True, если операнд равен False и наоборот.

```python
# Оператор == (равно)
a = 5
b = 5
result_1 = a == b
print("a == b:", result_1)  # True, так как 5 равно 5

# Оператор != (не равно)
a = 10
b = 5
result_2 = a != b
print("a != b:", result_2)  # True, так как 10 не равно 5

# Оператор > (больше)
a = 7
b = 3
result_3 = a > b
print("a > b:", result_3)  # True, так как 7 больше 3

# Оператор < (меньше)
a = 2
b = 8
result_4 = a < b
print("a < b:", result_4)  # True, так как 2 меньше 8

# Оператор >= (больше или равно)
a = 5
b = 5
result_5 = a >= b
print("a >= b:", result_5)  # True, так как 5 больше или равно 5

# Оператор <= (меньше или равно)
a = 3
b = 4
result_6 = a <= b
print("a <= b:", result_6)  # True, так как 3 меньше или равно 4

# Пример с равенством строк
str1 = "apple"
str2 = "apple"
result_7 = str1 == str2
print("str1 == str2:", result_7)  # True, так как строки идентичны

# Пример с неравенством строк
str1 = "apple"
str2 = "banana"
result_8 = str1 != str2
print("str1 != str2:", result_8)  # True, так как 'apple' не равно 'banana'

# Пример сравнения строк (лексикографический порядок)
str1 = "apple"
str2 = "banana"
result_9 = str1 < str2
print("str1 < str2:", result_9)  # True, так как 'a' меньше 'b' в алфавите

# Пример сравнения строк (лексикографический порядок)
str1 = "dog"
str2 = "cat"
result_10 = str1 > str2
print("str1 > str2:", result_10)  # True, так как 'd' больше 'c' в алфавите
```

# Оператор and возвращает True, если оба условия истинны. Если хотя бы одно из условий ложно, возвращается False.

```python
age = 25
has_license = True

if age >= 18 and has_license:
    print("Вы можете водить автомобиль.")
else:
    print("Вы не можете водить автомобиль.")
```

# Оператор or возвращает True, если хотя бы одно из условий истинно. Если оба условия ложны, то результат будет False.#

image.png

```python
has_ticket = False
is_vip = True

if has_ticket or is_vip:
    print("Вы можете войти на концерт.")
else:
    print("Вы не можете войти на концерт.")
```
