```python
def factorial(n):
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res

for i in range(1, 6):
    print(i, '! = ', factorial(i), sep='')
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Пример использования
print(factorial(5))  # Вывод: 120
```

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Пример использования
print(fibonacci(6))  # Вывод: 8
```

Задача 1: Вычисление степени числа
Напишите рекурсивную функцию для вычисления степени числа (например, 2^3 = 8).

```python
def power(base, exp):
    if exp == 0:
        return 1
    else:
        return base * power(base, exp - 1)

# Пример использования
print(power(2, 3))  # Вывод: 8
print(power(5, 2))  # Вывод: 25
```
