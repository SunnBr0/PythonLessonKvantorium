```python
import tkinter as tk
from tkinter import messagebox

# Функции для обработки событий
def say_hello():
    """Функция для вывода приветствия"""
    name = entry_name.get()
    if name:
        messagebox.showinfo("Приветствие", f"Привет, {name}!")
    else:
        messagebox.showwarning("Предупреждение", "Введите имя!")

def clear_text():
    """Функция для очистки текстового поля"""
    entry_name.delete(0, tk.END)

# Создание главного окна
root = tk.Tk()
root.title("Простое Tkinter Приложение")
root.geometry("400x300")

# Метка для имени пользователя
label_user = tk.Label(root, text="Введите ваше имя:")
label_user.pack(pady=10)

# Текстовое поле для ввода имени
entry_name = tk.Entry(root, width=30)
entry_name.pack()

# Кнопка 1: Приветствие
button1 = tk.Button(root, text="Привет!", padx=10, pady=5, command=say_hello)
button1.pack(pady=10)

# Кнопка 2: Очистить текстовое поле
button2 = tk.Button(root, text="Очистить", padx=10, pady=5, command=clear_text)
button2.pack(pady=5)

# Дополнительная метка
label_info = tk.Label(root, text="Нажмите 'Привет!' чтобы поприветствовать.")
label_info.pack(pady=10)

# Запуск основного цикла
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, filedialog

class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Универсальное Tkinter Приложение")
        self.root.geometry("600x400")

        # Создаем меню
        self.menu_bar = tk.Menu(self.root)
        self.root.config(menu=self.menu_bar)

        # Меню "Файл"
        self.file_menu = tk.Menu(self.menu_bar, tearoff=0)
        self.file_menu.add_command(label="Открыть", command=self.open_file)
        self.file_menu.add_command(label="Сохранить", command=self.save_file)
        self.file_menu.add_separator()
        self.file_menu.add_command(label="Выход", command=self.root.quit)
        self.menu_bar.add_cascade(label="Файл", menu=self.file_menu)

        # Меню "Помощь"
        self.help_menu = tk.Menu(self.menu_bar, tearoff=0)
        self.help_menu.add_command(label="О программе", command=self.about)
        self.menu_bar.add_cascade(label="Помощь", menu=self.help_menu)

        # Основной фрейм
        self.main_frame = ttk.Frame(self.root, padding="10")
        self.main_frame.pack(fill=tk.BOTH, expand=True)

        # Метка
        self.label = ttk.Label(self.main_frame, text="Добро пожаловать!", font=("Arial", 14))
        self.label.pack(pady=10)

        # Текстовое поле
        self.text_entry = ttk.Entry(self.main_frame, width=50)
        self.text_entry.pack(pady=5)

        # Кнопка
        self.button = ttk.Button(self.main_frame, text="Нажми меня", command=self.on_button_click)
        self.button.pack(pady=10)

        # Чекбокс
        self.checkbox_var = tk.BooleanVar()
        self.checkbox = ttk.Checkbutton(self.main_frame, text="Подтвердить", variable=self.checkbox_var)
        self.checkbox.pack(pady=5)

        # Радиокнопки
        self.radio_var = tk.StringVar(value="option1")
        self.radio1 = ttk.Radiobutton(self.main_frame, text="Опция 1", value="option1", variable=self.radio_var)
        self.radio2 = ttk.Radiobutton(self.main_frame, text="Опция 2", value="option2", variable=self.radio_var)
        self.radio1.pack()
        self.radio2.pack()

        # Список
        self.listbox = tk.Listbox(self.main_frame, height=5)
        self.listbox.pack(pady=10)
        for item in ["Элемент 1", "Элемент 2", "Элемент 3"]:
            self.listbox.insert(tk.END, item)

        # Прогрессбар
        self.progress = ttk.Progressbar(self.main_frame, orient="horizontal", length=200, mode="determinate")
        self.progress.pack(pady=10)
        self.progress["value"] = 50  # Устанавливаем значение прогресса

    def on_button_click(self):
        """Обработчик нажатия кнопки"""
        text = self.text_entry.get()
        if text:
            messagebox.showinfo("Информация", f"Вы ввели: {text}")
        else:
            messagebox.showwarning("Предупреждение", "Поле ввода пустое!")

    def open_file(self):
        """Открытие файла"""
        file_path = filedialog.askopenfilename(
            title="Выберите файл",
            filetypes=(("Текстовые файлы", "*.txt"), ("Все файлы", "*.*"))
        )
        if file_path:
            messagebox.showinfo("Файл открыт", f"Выбран файл: {file_path}")

    def save_file(self):
        """Сохранение файла"""
        file_path = filedialog.asksaveasfilename(
            title="Сохранить файл",
            defaultextension=".txt",
            filetypes=(("Текстовые файлы", "*.txt"), ("Все файлы", "*.*"))
        )
        if file_path:
            with open(file_path, "w") as file:
                file.write("Пример текста")
            messagebox.showinfo("Файл сохранен", f"Файл сохранен: {file_path}")

    def about(self):
        """О программе"""
        messagebox.showinfo("О программе", "Это универсальное приложение на Tkinter.")

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

# Создаем главное окно
root = tk.Tk()
root.title("Простая программа с tkinter")
root.geometry("300x200")

# Создаем стиль
style = ttk.Style()
style.configure("TButton", padding=6,, background="#0059b3", foreground="white")
style.configure("TLabel", padding=6, relief="", background="#0059b3", foreground="white")

# Функция для изменения стиля при нажатии на кнопку
def change_style():
    style.configure("TButton", background="#ff6347", foreground="white")
    style.configure("TLabel", background="#ff6347", foreground="white")

# Создаем метку
label = ttk.Label(root, text="Привет, Tkinter!", style="TLabel")
label.pack(pady=20)

# Создаем кнопку
button = ttk.Button(root, text="Изменить стиль", style="TButton", command=change_style)
button.pack(pady=20)

# Запускаем главный цикл обработки событий
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import random

# Список слов для игры
words = ["банан", "яблоко", "вишня"]

def shuffle_word(word):
    # Перемешиваем буквы в слове
    word_list = list(word)
    random.shuffle(word_list)
    return ''.join(word_list)

def check_guess():
    guess = entry.get().lower()
    if guess == current_word:
        messagebox.showinfo("Победа!", "Правильно! Вы угадали слово!")
        reset_game()
    else:
        messagebox.showerror("Неправильно", "Попробуйте еще раз!")

def reset_game():
    global current_word
    current_word = random.choice(words)
    shuffled = shuffle_word(current_word)
    word_label.config(text=shuffled)
    entry.delete(0, tk.END)

# Создаем основное окно
root = tk.Tk()
root.title("Словоместка")

# Выбираем случайное слово и перемешиваем его
current_word = random.choice(words)
shuffled_word = shuffle_word(current_word)

# Создаем виджеты
word_label = tk.Label(root, text=shuffled_word, font=("Helvetica", 24))
entry = tk.Entry(root, font=("Helvetica", 18))
check_button = tk.Button(root, text="Проверить", command=check_guess, font=("Helvetica", 16))
reset_button = tk.Button(root, text="Новое слово", command=reset_game, font=("Helvetica", 16))
word_label.pack(pady=20)
entry.pack(pady=10)
check_button.pack(pady=10)
reset_button.pack(pady=10)

# Запускаем главный цикл
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

# Создаем главное окно
root = tk.Tk()
root.title("Стиль через ttk")
root.geometry("400x300")

# Создаем объект стиля
style = ttk.Style()

# Настройка стиля для кнопки
style.configure(
    "Custom.TButton",  # Имя стиля
    background="lightgreen",  # Цвет фона
    foreground="black",       # Цвет текста
    font=("Verdana", 14),     # Шрифт
    padding=10                # Отступы
)

# Создаем кнопку с применением стиля
button = ttk.Button(
    root,
    text="Тематическая кнопка",
    style="Custom.TButton"  # Применяем стиль
)
button.pack(pady=20)

# Настройка стиля для метки
style.configure(
    "Custom.TLabel",
    background="pink",
    foreground="blue",
    font=("Times New Roman", 16, "italic"),
    padding=5

)

# Создаем метку с применением стиля
label = ttk.Label(
    root,
    text="Тематическая метка",
    style="Custom.TLabel"
)
label.pack(pady=10)

# Запускаем главный цикл
root.mainloop()
```
