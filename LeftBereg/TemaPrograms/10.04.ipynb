```python

import tkinter as tk
from tkinter import ttk
root = tk.Tk()
root.title("Стиль через ttk")
root.geometry("400x300")
style = ttk.Style()
style.configure(
    "Custom.TButton",
    background="lightgreen",
    foreground = "red",
    font=("Verdana", 14),
    padding=10
)
buttons = ttk.Button(
    root,
    text="Тематическая кнопка",
    style="Custom.TButton"
)

styleEntry = ttk.Style()
styleEntry.configure(
    "Custom.TEntry",
    fieldbackground="lightgray", 
    foreground="green",
    insertcolor="red", 
    insertwidth=3, 
    font=("Arial", 14)
)
entry = ttk.Entry(root,style="Custom.TEntry")
entry.pack(pady=20)
buttons.pack(pady=20)
root.mainloop()
```

```python
import tkinter as tk 
root = tk.Tk()
root.title("Пример")
root.geometry("400x300")
mainMenu = tk.Menu(root)
root.config(menu=mainMenu)

fileMenu = tk.Menu(mainMenu,tearoff=0)
fileMenu.add_command(label="Открыть",command=lambda: print("Открыть файл"))
fileMenu.add_command(label="Сохранить",command=lambda: print("Сохранить файл"))
exportMenu = tk.Menu(fileMenu,tearoff=0)
exportMenu.add_command(label="В PDF",command=print("Экспорт в PDF"))
exportMenu.add_command(label="В PNG",command=print("Экспорт в PNG"))

fileMenu.add_cascade(label="Экспорт",menu=exportMenu)
fileMenu.add_separator()
fileMenu.add_command(label="Выход",command=root.quit)
mainMenu.add_cascade(label="Файл",menu=fileMenu)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, filedialog

def open_file():
    """Открытие файла"""
    file_path = filedialog.askopenfilename(
        title="Выберите файл",
        filetypes=(("Текстовые файлы", "*.txt"), ("Все файлы", "*.*"))
    )
    if file_path:
        messagebox.showinfo("Файл открыт", f"Выбран файл: {file_path}")

def save_file():
    """Сохранение файла"""
    file_path = filedialog.asksaveasfilename(
        title="Сохранить файл",
        defaultextension=".txt",
        filetypes=(("Текстовые файлы", "*.txt"), ("Все файлы", "*.*"))
    )
    if file_path:
        with open(file_path, "w") as file:
            file.write("Пример текста")
        messagebox.showinfo("Файл сохранен", f"Файл сохранен: {file_path}")

def about():
    """О программе"""
    messagebox.showinfo("О программе", "Это универсальное приложение на Tkinter.")

# Создаем главное окно
root = tk.Tk()
root.title("Универсальное Tkinter Приложение")
root.geometry("600x400")
# Создаем меню
menu_bar = tk.Menu(root)
root.config(menu=menu_bar)
# Меню "Файл"
file_menu = tk.Menu(menu_bar, tearoff=0)
file_menu.add_command(label="Открыть", command=open_file)
file_menu.add_command(label="Сохранить", command=save_file)
file_menu.add_separator()
file_menu.add_command(label="Выход", command=root.quit)
menu_bar.add_cascade(label="Файл", menu=file_menu)
# Меню "Помощь"
help_menu = tk.Menu(menu_bar, tearoff=0)
help_menu.add_command(label="О программе", command=about)
menu_bar.add_cascade(label="Помощь", menu=help_menu)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

root = tk.Tk()
root.title("Расширенная стилизация")
root.geometry("400x300")

# Создаем объект стиля
style = ttk.Style()

# Настройка стиля для кнопки
style.configure(
    "Custom.TButton",
    background="lightgreen",
    foreground="red",  # Цвет текста
    font=("Verdana", 14, "italic"),  # Шрифт
    padding=10,  # Отступы
    relief="raised",  # Стиль границы
    anchor="center"  # Выравнивание текста
)

# Динамическое изменение стиля при наведении курсора
style.map(
    "Custom.TButton",
    background=[("active", "lightblue"), ("pressed", "darkblue")],  # Изменение цвета фона
    foreground=[("active", "white"), ("pressed", "yellow")]  # Изменение цвета текста
)

# Создаем кнопку с применением стиля
button = ttk.Button(
    root,
    text="Тематическая кнопка",
    style="Custom.TButton"
)
button.pack(pady=20)

# Запускаем главный цикл
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

root = tk.Tk()
root.title("Стилизация текстового поля")
root.geometry("400x300")

# Создаем объект стиля
style = ttk.Style()

# Настройка стиля для текстового поля
style.configure(
    "Custom.TEntry",
    fieldbackground="lightgray",  # Цвет фона
    foreground="green",  # Цвет текста
    insertcolor="red",  # Цвет курсора
    insertwidth=3,  # Толщина курсора
    font=("Arial", 14)
)

# Создаем текстовое поле с применением стиля
entry = ttk.Entry(
    root,
    style="Custom.TEntry"
)
entry.pack(pady=20)

# Запускаем главный цикл
root.mainloop()
```
