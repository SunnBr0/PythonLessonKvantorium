```python
import matplotlib.pyplot as plt
dictPeople = {"Дети":130,"Взрослые":50,"Пожилые":20}
valueCount = dictPeople.values()
valueType = dictPeople.keys()
plt.pie(valueCount,labels=valueType,autopct="%1.1f%%")
plt.legend()
plt.title("Страна")
plt.xlabel("Чиссленость населения")
plt.ylabel("Категории")
plt.show()
```
