```python
import matplotlib.pyplot as plt
x = ["","","",""]
y = [10,15,30,-25]
plt.plot(x,y,color="red",marker="o",markersize=8)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["","","",""]
y = [10,15,30,-25]
plt.bar(x,y,label=" ",color="#1dabe2",alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["","","",""]
y = [10,15,30,-25]
plt.plot(x,y,color="red",marker="o",markersize=8)
plt.bar(x,y,label=" ",color="#1dabe2",alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
dictPeople = {"":130,"":50,"":20}
valueCount = dictPeople.values()
valueType = dictPeople.keys()
plt.pie(valueCount,labels=valueType,autopct="%1.1f%%")
plt.legend()
plt.title("")
plt.xlabel(" ")
plt.ylabel("")
plt.show()
```
