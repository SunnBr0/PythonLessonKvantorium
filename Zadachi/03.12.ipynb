```python
nameTovar = input("Введите название товара: ")
namePrice = int(input("Введите цену: "))
massTovar = int(input("Введите вес товара: "))
cash = int(input("Деньги : "))
print("Чек")
print(f"{nameTovar} - {massTovar}кг - {namePrice}руб/кг")
end = namePrice * massTovar
print(f"Итого: {end}руб")
print(f"Внесено: {cash}руб")
cashback = cash - end
if cash < end:
    print("Недостаточно")
else:
    print(f"Сдача: {cashback}руб")
```

```python
nameTovar = input("Название товара: ")
priceTovar = int(input("Цена товара: "))
massTovar = int(input("Масса товара: "))
money = int(input("Кол-во денег: "))
print("Чек")
print(nameTovar,"-",massTovar,"кг","-",priceTovar,"руб/кг")
print("Итого: ",priceTovar*massTovar,"руб")
print("Внесено:",money,"руб")
if money < massTovar*priceTovar:
    print("денег не хватает,берём кредит")
else:
    print("Сдача:",money - massTovar*priceTovar,"руб")
```
