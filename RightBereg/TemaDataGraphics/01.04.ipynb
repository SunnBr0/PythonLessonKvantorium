```python
import matplotlib.pyplot as plt
x = ["Январь","Февраль","Март","Апрель"]
y = [5,-15,-3,15]
plt.title("Величина температуры")
plt.plot(x,y,color="red",marker="^",markersize=8)
plt.bar(x,y,color="#8aea15",alpha=0.3)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
dictPeople = {"Дети":30,"Взрослые":50,"Пожилые":20}
valueCount = dictPeople.values()
valueType = dictPeople.keys()
plt.title("Величина температуры")
plt.pie(valueCount,labels=valueType)
plt.show()
```
