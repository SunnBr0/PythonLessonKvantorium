```python
import matplotlib.pyplot as plt
x = [12,13,123,1,-1,0]
y = [-2,12,1,4,5,0]
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["","","",""]
y = [5,-15,-3,15]
plt.plot(x,y,color="red",marker="^",markersize=8)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["","","",""]
y = [5,-15,-3,15]
plt.title(" ")
plt.bar(x,y,color="#8aea15",alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["","","",""]
y = [5,-15,-3,15]
plt.title(" ")
plt.plot(x,y,color="red",marker="^",markersize=8)
plt.bar(x,y,color="#8aea15",alpha=0.3)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
dictPeople = {"":30,"":50,"":20}
valueCount = dictPeople.values()
valueType = dictPeople.keys()
plt.title(" ")
plt.pie(valueCount,labels=valueType)
plt.show()
```
